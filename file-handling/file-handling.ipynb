```python
# The 'with' keyword is beneficial because it ensures that the file is properly closed after its suite finishes,
# even if an exception is raised. This is a good practice to prevent resource leaks.
with open("abc.txt") as file:
    print(file.read())

print("pakistan zindabad")    
```

```python
# readline()
with open("abc.txt") as file:
    print(file.readline()) # read only one line (first line) 
    print(file.readline()) # read only one line (second line)
```

```python
# readline()
with open("abc.txt") as file:
    print(file.readline() , end=" ")  # end=" " ensures that the next print statement continues on the same line with a space
    print(file.readline() , end=" ") # end=" " removed \n
```

```python
# readlines()
with open("abc.txt") as file:
    print(file.readlines())  # readlines() reads all the lines in a file and returns them as a list of strings.
                             # we can applay here all list  method
                             # ['hello world\n', 'i am a programmer\n', 'line1\n', 'line2\n', 'line3\n', 'line4\n', 'line5']
```

```python
# readlines()
with open("abc.txt") as file:
    print(file.readlines()[ : 3])  # we can do silicing due to list
```

```python
# Open and read the contents of 'abc1.txt' file using the 'with' statement
with open('abc1.txt' , "r") as file1:
    print(file1.read())
```

```python

with open('abc1.txt' , "w") as file1:
    print(file1.write("abc")) # This will remove all data from 'abc1.txt' file and then write "abc" to it
```

```python

with open('abc1.txt' , "r+") as file1:
    print(file1.write("xyz")) # This will remove all data from 'abc1.txt' file because the cursor is at the first index 
                              # when the file is opened in 'r+' mode. Therefore, it removes all data and writes "xyz".
    print(file1.read() , end=" ")  # return empty space beacuse now cursor is end of sentence 
    file1.seek(0)    # back cursor 0 index 
    print(file1.read() , end=" ")  # now read all data and return
                     
```

```python
with open("newfile" , "w") as file2 :
    print(file2.write("computer"))

# if file exist then overwrite the text
# if file not exist then it will create new file
```

```python
# This mode checks if the file exists. If it does, it appends data to the end of the file. 
# If the file does not exist, it creates a new file and starts appending data to it.
with open("newfile" , "a") as file3 :
    print(file3.write("laptop")) # add in end 
```

```python
# The 'x' mode is used for creating a new file and opening it for writing.
# If the file already exists, the operation will fail with a FileExistsError.
with open("newfile" , "x") as file3 :
    print(file3.write("laptop")) 
```

```python
 # The 'rb+' mode opens the file for both reading and writing in binary mode.
# The file pointer is placed at the beginning of the file.
with open("newfile" , "rb+") as file3 :
        print(file3.read())
     
```
