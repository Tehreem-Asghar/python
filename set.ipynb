```python

# set is a built-in Python data type that represents an unordered collection of unique elements.

my_set: set = {123, 452, 5, 6}
my_set2: set = set([123, 452, 5, 6])
print("my_set            = ", my_set)
print("my_set2           = ", my_set2)
print("type(my_set)      = ", type(my_set))
print("type(my_set2)     = ", type(my_set2))
print("my_set == my_set2 = ", my_set == my_set2)


# Holds only Immutable Objects

# my_set = {123, 452, 5, 6, [1, 2, 3]}  # TypeError: unhashable type: 'list' cant not add list to set beacuse list is mutable
# print("my_set = ", my_set) # TypeError: unhashable type: 'list'



# It can hold multiple data types at once.

my_set = {123, 452, 5, 6, "Hello", 3.14, (1, 2, 3)}
print("my_set = ", my_set) # {3.14, 5, 6, 'Hello', 123, 452, (1, 2, 3)}

# It can hold only immutable objects.

# The set is unordered
set2: set = {'Java', 'Python', 'JavaScript', 'java'}
print(set2)  # {'Java', 'Python', 'java', 'JavaScript'}

# The set is Unchangeable

# set2[0] = 'C++'  # TypeError: 'set' object does not support item assignment


# methods of set

# add()	Adds an element to the set
# remove() 
# clear()	Removes all the elements from the set
# copy()	Returns a copy of the set
# difference()	Returns a set containing the difference between two or more sets
# difference_update()	Removes the items in this set that are also included in another, specified set
# discard()	Remove the specified item
# intersection()	Returns a set, that is the intersection of two other sets
# intersection_update()	Removes the items in this set that are not present in other, specified set(s)


# add()	Adds an element to the set
my_set = {123, 452, 5, 6}
my_set.add(100)
print("my_set = ", my_set)  # {5, 6, 100, 452, 123}

# clear()	Removes all the elements from the set
my_set.clear()
print("my_set = ", my_set)  # set()


# copy()	Returns a copy of the set
my_set = {123, 452, 5, 6}
my_set2 = my_set.copy()
print("my_set = ", my_set)  # {5, 6, 123, 452}
print("my_set2 = ", my_set2)  # {5, 6, 123, 452}
print("my_set == my_set2 = ", my_set == my_set2)  # True

#difference_update()  remove multiple element at once.

my_set: set = {1, 2, 3, 4, 5, 'A', 'a'}
print("Before: my_set = ", my_set)
my_set.difference_update({1, 5, 3, 'A'})
print("After:  my_set = ", my_set)

# update() 
print("Before: ", my_set)
# Add multiple items 
my_set.update([7, 8, 9, "Hello"])
print(my_set)  # Output: {4, 5, 6, 7, 8, 9}




# Using the union() method or | operator:

# In Python, the union() method or the | operator is used to combine two sets into a single set. This operation returns a new set containing all unique elements from both sets.

# Using the union() method:

# The union() method is a built-in method of the set data type in Python. It takes an iterable (such as a set, list, or tuple) as an argument and returns a new set containing all unique elements from both the original set and the iterable.


my_set: set   = {1, 2, 3, 5}
my_set_2: set = {1, 5, 6, 7}

my_set3: set  = my_set.union(my_set_2)
print(my_set3) # Output: {1, 2, 3, 5, 6, 7}

# Using the | operator:
print(my_set | my_set_2 ) # Output: {1, 2, 3, 5, 6, 7}

# Using the intersection() method or & operator:

# In Python, the intersection() method or the & operator is used to find the common elements between two sets.
#  This operation returns a new set containing only the elements that are present in both sets.

print(my_set.intersection(my_set_2)) # Output: {1, 5}

# Using the & operator:
print(my_set & my_set_2) # Output: {1, 5}



print(my_set.discard(2)) # Output: None
print(my_set) # Output: {1, 3, 5}




my_set: set  = {1,2,3, "Hello! World", 4,5,6}
my_set2: set = {1,2,3, "Hello! World", 8,9}

print("difference()           = ", my_set.difference(my_set2)) #Returns a set containing the difference between two or more sets
print("intersection()         = ", my_set.intersection(my_set2))#Return a set that contains the items that exist in both set
print("union()                = ", my_set.union(my_set2))#Return a set that contains all items from both sets, duplicates are excluded:
print("symmetric_difference() = ", my_set.symmetric_difference(my_set2))#Return a set that contains only unique items from both sets

```

```python
# Frozenset 
# Python mai frozenset ek immutable (unchangeable) version hota hai set ka. 
# Ye set ki tarah unique elements store karta hai, lekin iske elements ko modify nahi kiya ja sakta.



# Normal Set (Mutable)
my_set = {1, 2, 3}
my_set.add(4)  # ✅ Allowed
print(my_set)  # Output: {1, 2, 3, 4}

# Frozenset (Immutable)
my_frozenset = frozenset([1, 2, 3])
# my_frozenset.add(4)  # ❌ Error: 'frozenset' object has no attribute 'add'



# Frozenset as a Dictionary Key
# Kyuki frozenset immutable hai, ye dictionary ka key ban sakta hai, lekin normal set nahi ban sakta.

# Using frozenset as dictionary key



my_dict = {
   frozenset({1, 2, 3}) : "Immutable Key"
}
print(my_dict)  # ✅ Output: {frozenset({1, 2, 3}): 'Immutable Key'}



# ❌ Agar normal set ko key banayenge to error aayega

# my_dict = {
#     {1, 2, 3}: "Mutable Key"  # ❌ TypeError: unhashable type: 'set'
# }
```
