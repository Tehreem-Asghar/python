```python
from typing import List, Dict
from datetime import date
import pandas as pd
import pandera as pa

# Defining schema for validation
fee_transaction_schema = pa.DataFrameSchema({
    "date": pa.Column(pa.DateTime),
    "fee": pa.Column(pa.Int)
})

# Defining fee transactions data
fee_transactions: List[Dict[str, object]] = [
    {"date": date(2023, 1, 10), "fee": 48000},
    {"date": date(2023, 2, 15), "fee": 51000},
    {"date": date(2023, 3, 20), "fee": 53000},
    {"date": date(2023, 4, 5), "fee": 47000},
    {"date": date(2023, 5, 12), "fee": 50000},
    {"date": date(2023, 6, 18), "fee": 52000},
    {"date": date(2023, 7, 22), "fee": 65000},
    {"date": date(2023, 8, 14), "fee": 70000},
    {"date": date(2023, 9, 30), "fee": 58000},
    {"date": date(2023, 10, 25), "fee": 55000},
]

# Creating DataFrame
df_fee_transactions = pd.DataFrame(fee_transactions)
df_fee_transactions["date"] = pd.to_datetime(df_fee_transactions["date"])  # Convert date column to datetime

df_fee_transactions = fee_transaction_schema.validate(df_fee_transactions)

# Display DataFrame
print(df_fee_transactions, '\n============================================================')
print(df_fee_transactions.dtypes)
```

# Short Definition of pd.cut() 🔥
* pd.cut() continuous numerical data ko bins (categories) me divide karne ke liye use hota hai, taake values ko meaningful groups me classify kiya ja sake. 🚀

* 💡 Example: pd.cut([10, 25, 40], bins=[0, 20, 50], labels=["Low", "High"])

* 👉 Output: ["Low", "High", "High"]


```python
import pandas as pd

# Fee amounts
fee_data = {'fee': [200, 800, 3000, 4500, 500, 1200, 3800]}
df = pd.DataFrame(fee_data)

# Fee ke liye bins define kiye
bins = [0, 500, 1000, 4000, 5000]
labels = ["Low", "Medium", "High", "Very High"]

# pd.cut() ka use karke fee categorize ki
df['Fee Category'] = pd.cut(df['fee'], bins=bins, labels=labels)

print(df)
```

```python
# Fee ko bins me categorize karna aur count karna
fee_bins = pd.cut(df_fee_transactions['fee'], bins=[1, 300, 500, 1000, 4000, 5000]).value_counts()

# Output print karna
print(fee_bins)
```

```python
# Fee ko bins me categorize karna aur count karna
fee_bins = pd.cut(df_fee_transactions['fee'], bins=[1, 300, 500, 1000, 4000, 5000]).value_counts(normalize=False)*100

# Output print karna
print(fee_bins)
```

# pd.qcut() Simple Explanation ✅
* pd.qcut() data ko equal-sized groups me divide karta hai. Ye quantiles (percentiles) ko use karta hai taake har group me barabar elements ho.

```python
# Example:
# Agar humare paas 6 numbers hain aur unko 3 groups me divide karna hai:

import pandas as pd

data = [10, 20, 30, 40, 50, 60]  # Some numbers
bins = pd.qcut(data, q=3, labels=["Low", "Medium", "High"])
print(bins)
```

# Group by

* dataframe.groupby(['column1','column2']) Ye data ko multiple columns ke basis par group karne ke liye use hota hai.
Iska output grouped object hota hai jo further operations ke liye ready hota hai.

* dataframe.groupby(['column1','column2'])[['col1','col2'].aggregate([min,max]) Ye har group ka minimum aur maximum value calculate karne ke liye use hota hai.

* dataframe.groupby(['column1','column2']).apply(function())  Ye har group par ek custom function apply karne ke liye use hota hai.




* groupby() DataFrame me similar values wale rows ko group karne ke liye use hota hai.

* Kya Karta Hai?
* Data ko categories me divide karta hai.
* Har group ke summary statistics (sum, count, mean, etc.) calculate karne me madad karta hai.


* Jab hume kisi specific category ya group ke basis par analysis karni ho, jaise:

* Students ke marks ka average class-wise
* Har city me kitne log hain
* Har product ka total sales

```python
import pandas as pd

# Students DataFrame
students = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 5],
    'name': ['Ali', 'Sara', 'Ahmed', 'Zainab', 'Hassan'],
    'teacher_id': [101, 102, 103, 101, 102]
})

# Teachers DataFrame
teachers = pd.DataFrame({
    'teacher_id': [101, 102, 103],
    'teacher_name': ['Mr. Khan', 'Ms. Fatima', 'Dr. Ahmed']
})

# Student Fees DataFrame
student_fees = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 5],
    'fee_paid': [50000, 52000, 49000, 53000, 51000]
})

# Display the DataFrames
print("Students DataFrame:\n", students, "\n======================================================")
print("Teachers DataFrame:\n", teachers, "\n======================================================")
print("Student Fees DataFrame:\n", student_fees, "\n=======================================================")
```
