# apply custom function on multiple columns and return on column or more than one
* loc
  * dataframe[rows , columns]
        * exact labels value
* iloc
  * dataframe[rows,columns]
         * numerical indexing value
* at
   * cell value with exact label
* iat    
   * cell value with numerical index

```python
print(ss1[0:10],'\n===============================================================================') # direct silincing apply on rows index
print(ss1.iloc[1:11],'\n===============================================================================') # numpy silicing apply on columns and rows index
print(ss1.iloc[1:11 , 1:3],'\n===============================================================================')
print(ss1.loc[1:11],'\n===============================================================================') #
print(ss1.loc[1:11 , "s1":"s3"],'\n===============================================================================') # direct silincing apply on  columns and rows labels
print(ss1.at[0,"s1" ],'\n===============================================================================') # specific value
print(ss1.iat[0,0],'\n===============================================================================') # specific value
```

# .apply()
Pandas ka .apply() method har row ya column par ek function apply karne ke liye use hota hai. Yeh element-wise operation perform karne ke liye kaafi flexible hai.

```python
import pandas as pd
import numpy as np
from typing import Any

# Sample DataFrame with student names and marks
data: dict[str, list[Any]] = {
    "Name": ["Ali", "Sana", "Hamza", "Ayesha", "Bilal"],
    "Maths": [85, 78, 92, 88, 76],
    "Science": [90, 82, 95, 85, 79],
    "English": [88, 76, 89, 84, 80]
}

dframe: pd.DataFrame = pd.DataFrame(data)
# Function to calculate average marks
def calculate_average(row: pd.Series) -> float:
    return (row["Maths"] + row["Science"] + row["English"]) / 3

# Apply function to calculate the average marks for each student
dframe["Average Marks"] = dframe.apply(calculate_average, axis=1)

# Function to assign grade based on average marks
def assign_grade(avg_marks: float) -> str:
    if avg_marks >= 90:
        return "A+"
    elif avg_marks >= 80:
        return "A"
    elif avg_marks >= 70:
        return "B"
    else:
        return "C"

# Apply function to assign grades
dframe["Grade"] = dframe["Average Marks"].apply(assign_grade)

# Display DataFrame
print(dframe)

# DataFrame Summary
print("\nData Types:\n", df.dtypes)
print("\nSummary:\n", df.describe())
```
