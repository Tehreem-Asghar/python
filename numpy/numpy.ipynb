#  numpy.in1d()

* numpy.in1d() dono arrays ka comparison karta hai aur check karta hai ke pehle array ke elements dusre array me hain ya nahi.

* Iska output Boolean array hota hai jo batata hai ke kaunse elements present hain aur kaunse nahi.



```python
import numpy as np

a = np.array([1, 2, 3, 4, 5])  # Array1
b = np.array([3, 4, 7, 8])     # Array2

result = np.in1d(a, b) # check if elements of Array1 are present in Array2
print(result)  # Output: [False, False,  True,  True, False]
```

```python
result = np.in1d(a, b) # check if elements of Array1 are present in Array2
print(a[result]) # a[] karna sa true value print karega
```

```python
# Agar tumhe sirf common elements chahiye, to in1d() ko indexing ke saath use kar sakte hai.
common_elements = a[np.in1d(a, b)]
print(common_elements)
```

# numpy.where()
* numpy.where() ek conditional function hai jo kisi condition ke basis pe array ke elements ko filter ya replace karta hai.
* Agar condition True hoti hai, to us index ka value return ya replace karta hai.


numpy.where(condition, x, y)

condition → Jo check karni hai (True ya False).

x → Jo value return hogi agar condition True ho.

y → Jo value return hogi agar condition False ho.

Agar sirf condition dete ho, to True wale indexes return hotay hain.

```python
#  Indexes Find Karna

import numpy as np

a = np.array([10, 20, 30, 40, 50])

indexes = np.where(a > 25)
print(indexes)
```

```python
# Values Replace Karna

b = np.array([10, 25, 30, 15, 50])

new_array = np.where(b > 20, 1, 0)  # Agar value > 20 hai to 1, warna 0
print(new_array)
```

```python
# Multiple Conditions

c = np.array([5, 15, 25, 35, 45])

result = np.where((c > 10) & (c < 40), "Inside", "Outside")
print(result)
```

```python
#  2D Array Ke Saath Use

d = np.array([[1, 2, 3], [4, 5, 6]])

result = np.where(d % 2 == 0, "Even", "Odd")
print(result)
```

# np.arange() + reshape() ka Use
* np.arange(n)

arange() ek sequence of numbers generate karta hai 0 se lekar (n-1) tak.
Yahaan n ko manually calculate kar rahe hain.
* .reshape(x, y, z)

Iska matlab hai ke hum is array ko ek 3D shape dena chahte hain.
(x, y, z) ka matlab hai:
x = Number of matrices (depth)
y = Number of rows in each matrix
z = Number of columns in each matrix

```python
# shortcut to create 3D array                             2d,3rows,3coloumn  
c : NDArray[Shape["Size, Size , Size"] , Any] = np.arange(2*3*3).reshape(2,3,3)
                                                       #  3d,3rows,4coloumn 
d : NDArray[Shape["Size, Size , Size"] , Any] = np.arange(3*3*4).reshape(3,3,4)
print(c)
print("=========================")
print(d) 
```
